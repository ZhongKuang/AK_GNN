```python
import networkx as nx
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl 
%matplotlib inline
plt.rcParams['font.sans-serif']=['SimHei'] # 显示中文
plt.rcParams['axes.unicode_minus'] = False # 显示负号
```

### 创建无节点、无连接的空图

```python
G = nx.Graph()
```

```python
G
```

```python
G.nodes
```

```python
nx.draw(G)
```

### 添加单个节点

```python
G.add_node('刘备')
G.add_node('Tommy')
G.nodes
```

### 添加多个节点

```python
G.add_nodes_from(['诸葛亮','刘备'])
G.nodes
```

### 添加带属性特征的节点

```python
G.add_nodes_from([
    ('关羽',{'武器':'刀','武力':90,'智力':80}),
    ('张飞',{'武器':'矛','武力':85,'智力':75}),
    ('吕布',{'武器':'戟','武力':100,'智力':70})
])
G.nodes
```

```python
nx.draw(G)
```

### 创建一个首尾相连成串的PathGraph

```python
H = nx.path_graph(10)
nx.draw(H)
```

### 将H的节点添加到G中


```python
G.add_nodes_from(H)
G.nodes
```

```python
nx.draw(G)
```

### 将H作为一个节点添加到G中

```python
G.add_node(H)
G.nodes
```

```python
nx.draw(G)
```

### 小贴士

节点可以为任意可哈希的对象，比如字符串、图像、XML对象、甚至另一个Graph、自定义的节点对象
通过这种方式，你可以根据你的应用，自由灵活的构建：图为节点、文件为节点、函数为节点，等灵活的形式
